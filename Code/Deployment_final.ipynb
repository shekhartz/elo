```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error
from sklearn import model_selection, preprocessing, metrics
from pandas import DataFrame
from tensorflow.keras.models import Model, load_model
import pickle
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Case Study 1/Data/Dataset/train_FE2.csv')
```

```python
def predict(cardID):
  card_details = data.loc[data['card_id'] == cardID]
  y = card_details['target'].values
  x = card_details.drop(['card_id','first_active_month','target','outliers'], axis=1)

  model_path = '/content/drive/MyDrive/Colab Notebooks/Case Study 1/Data/Model/'
  model = pickle.load(open(model_path + 'lgb_kfold_model.sav', 'rb'))
  y_pred = model.predict(x)
  rmse = mean_squared_error(y_pred, y)**0.5
  return y, y_pred, rmse
```
